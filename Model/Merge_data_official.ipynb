```python
new_data = pd.read_csv('/content/drive/MyDrive/DS560/Data/Clean_data/fake_data1.csv')
new_data
```

```python
import numpy as np
new_data['Price'] = np.log10(new_data['Price'])
```

```python
new_data
```

Train-test split

```python
# train: 2010-2020
# test: 2021
# pred: 2022 1-6

train_data = new_data[new_data['Date']<='2020-12']
test_data = new_data[new_data['Date']>'2020-12']
print(len(train_data), len(test_data))
```

```python

train_data.head()
```

```python
y_train = train_data['Price']
X_train = train_data.drop(['Price','Date'], axis=1)
y_test = test_data['Price']
X_test = test_data.drop(['Price','Date'], axis=1)
```

```python

X_train.head()
```

```python
# onehot encoder
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
county = list(set(new_data['County']))
#enc = OneHotEncoder(handle_unknown='ignore')
#enc_county = pd.DataFrame(enc.fit_transform(county[['County']]).toarray())
```

```python
le = LabelEncoder()
le.fit(county)
enc_train_county = le.transform(X_train['County'])
enc_test_county = le.transform(X_test['County'])
# le.inverse_transform([0, 0, 1, 2])
```

```python
print(len(enc_train_county), len(enc_test_county))
```

```python
X_train['enc_county'] = enc_train_county
X_test['enc_county'] = enc_test_county
```

```python
X_train = X_train.drop(['County'], axis=1)
X_test = X_test.drop(['County'], axis=1)
```

```python
X_train.to_csv('/content/drive/MyDrive/DS560/Data/Clean_data/X_train1.csv')
X_test.to_csv('/content/drive/MyDrive/DS560/Data/Clean_data/X_test1.csv')
y_train.to_csv('/content/drive/MyDrive/DS560/Data/Clean_data/y_train1.csv')
y_test.to_csv('/content/drive/MyDrive/DS560/Data/Clean_data/y_test1.csv')
```
